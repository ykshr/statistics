# Central limit theorem

## What is the Central limit theorem?

The central limit theorem states that if you repeat the same trial a large number of times, the mean of your trial will converge to a normal distribution with mean $\mu$ and variance $\sigma^2/n$. The $\mu$ is the mean of the population, and $\sigma$ is the variance of the population. Interestingly enough, this theorem works whatever the population distribution is.

## Example

Suppose we operate a contact centre that receives an average of 10 calls every 5 minutes. If the calls follow a Poisson distribution, The probability of the number of calls is

$$
P(x)=e^{-10}\frac{10^x}{x!}
$$

The following graph shows that 9 and 10 calls have the highest possibility, and over 22 calls rarely occur.

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
from scipy import stats

l = 10
x = np.arange(0, 40)
y = stats.poisson.pmf(x, l)

plt.figure(figsize=(20, 10))
plt.plot(x, y)

plt.grid()
ax = plt.gca()
ax.xaxis.set_major_locator(ticker.MultipleLocator(1))

plt.title("Number of calls probability")
plt.xlabel("number of calls")
plt.ylabel("probability")

plt.show()
```

Let's generate the calls in 365-day and the averages of each day.

```python
from statistics import mean

averages = []
sample_size = 12 # one hour
hours = 24*365 # one year

for _ in range(hours):
    calls = poisson.rvs(l, size=sample_size)
    averages.append(mean(calls))

plt.figure(figsize=(20, 10))
plt.hist(averages, density=True, bins=np.unique(averages)+.5)

xx = np.linspace(min(averages), max(averages), 100)
plt.plot(xx, stats.norm.pdf(xx, loc=l, scale=l/sample_size))

plt.grid()
ax = plt.gca()
ax.xaxis.set_major_locator(ticker.MultipleLocator(1))

plt.title("Probability density function: average of calls")
plt.xlabel("number of calls")
plt.ylabel("probability")

plt.show()
```

The graph looks like normal distribution even though it may need further trials!

## Theorem

For the details: [Wikipedia](https://en.wikipedia.org/wiki/Central_limit_theorem)

Let $X_1, X_2, ..., X_n$ be a random sample from a population with mean $\mu$ and variance $\sigma^2$. Define $Z_n = \sqrt{x}(\bar{X}-\mu)/\sigma$. Then, for any $z\in\mathbb{R}$,

$$
P(Z_n < z) \rightarrow \Phi(z) \textrm{, as }n \rightarrow \infty
$$

where $\Phi(z)$ denotes the CDF of the N(0,1) distribution.
