# Exponential distribution

## What is exponential distribution?

An exponential distribution is a type of continuous probability distribution that describes the probability of a time between events. 

## Definition

Exponential distribution with mean $\mu$ is

$$
f(x) = \frac{1}{\mu}e^{-\frac{x}{\mu}} \ where \ x \geq 0
$$


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 5, 100)
pdf = np.exp(-x)
cdf = 1-np.exp(-x)

plt.plot(x, pdf, label="pdf")
plt.plot(x, cdf, label="cdf")
plt.legend()
plt.show()
```

Assume we have a web site where a new user comes every 1 minute on average ($\mu=1$). The probability that the next user will come less than 1 minute is the following.

```python
x = 1
cdf = 1-np.exp(-x)
print(cdf)
```

## Intuitive understanding of mean

Mean decides the probability mass centre of the distribution. So if we increase the mean, the probabilities of the larger time span increase.

```python
import matplotlib.pyplot as plt
import numpy as np

mues = [1, 2, 3]
for mu in mues:
    x = np.linspace(0, 5, 100)
    f = (1/mu)*np.exp(-x/mu)
    plt.plot(x, f, label='mu = '+str(mu))

plt.legend()
plt.show()
```

# Poisson distribution

## What is Poisson distribution?

A Poisson distribution is a type of discrete probability distribution that describes the probability of a given number of events.

## Definition

Poisson distribution with mean $\lambda$ is

$$
f(x) = e^{-\lambda}\frac{\lambda^k}{k!}
$$

This is the probability of an event occuring x times where the event occurs $\lambda$ times on average.

```python
import matplotlib.pyplot as plt
import numpy as np

lamb = 1
x = np.arange(0, 10)
pdf = np.exp(-lamb)*((lamb**x)/np.fromiter(map(np.math.factorial, x), dtype=np.int))
cdf = np.cumsum(pdf)

plt.plot(x, pdf, label="pdf")
plt.plot(x, cdf, label="cdf")
plt.legend()
plt.show()
```

## Intuitive understanding of mean

Mean decides the probability mass centre of the distribution. So if we increase the mean, of cause, the probabilities of the larger number of times an event occurs increase.

```python
import matplotlib.pyplot as plt
import numpy as np

lambs = [1, 2, 3, 4, 5, 6]
for lamb in lambs:
    x = np.arange(0, 10)
    f = np.exp(-lamb)*((lamb**x)/np.fromiter(map(np.math.factorial, x), dtype=np.int))
    plt.plot(x, f, label='lambda = '+str(lamb))

plt.legend()
plt.show()
```

# Relationship between Exponential distribution and Poisson distribution

The exponential distribution looks at the time span of the event. On the other hand, the Poisson distribution focuses on the number of times an event occurs.

Assume a significant earthquake hits our city every five years on average...
 - The exponential distribution describes the probability of the next earthquake occuring within two years. ($\mu=5$)
 - The Poisson distribution describes the probability of the number of times the earthquake occurs within two years. ($\lambda=2/5=0.4$)
