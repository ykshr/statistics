# Normal distribution

## What is normal distribution?

A normal distribution is a type of continuous and symmetrical probability distribution that mean has the highest probability. Since means of any distribution converge to normal distribution (central limit theorem), normal distribution is one of the most popular distribution.

## Definition

Normal distribution is mean: $\mu$ and variance: $\sigma^2$.

$$
f(x) = \frac{1}{\sqrt{2\pi\sigma^2}}\exp(-\frac{(x-\mu)^2}{2\sigma^2})
$$


Standard normal distribution is mean: $\mu=0$ and variance: $\sigma^2=1$.

$$
f(x) = \frac{1}{\sqrt{2\pi}}\exp(-\frac{x^2}{2})
$$


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 100)
f = (1/np.sqrt(2*np.pi))*np.exp(-x**2/2)

plt.plot(x, f)
plt.show()
```

## Intuitive understanding of mean and variance

### Mean

Mean decides the centre of the distribution. So if we change the mean, the distribution moves to the right or left.

```python
import matplotlib.pyplot as plt
import numpy as np

mues = [-2, 0, 2]
sigma = 1
for mu in mues:
    x = np.linspace(-5, 5, 100)
    f = (1/np.sqrt(2*np.pi*sigma**2))*np.exp(-(x-mu)**2/(2*sigma**2))

    plt.plot(x, f)
plt.show()
```

### Variance

Variance decides how much distribute the data. So if we change the variance, the distribution gets taller or lower.

```python
import matplotlib.pyplot as plt
import numpy as np

mu = 0
sigmas = [1, 2, 3]
for sigma in sigmas:
    x = np.linspace(-5, 5, 100)
    f = (1/np.sqrt(2*np.pi*sigma**2))*np.exp(-(x-mu)**2/(2*sigma**2))

    plt.plot(x, f)
plt.show()
```

## Origin of the normal distribution

To understand definition of the normal distribution, I want to create the function from scratch.

### Step1: Base function

At the normal distribution, we want to describe a probability distribution that has the highest probability at the middle and decrease symmetrically. The simplest function with this characteristic can be exponential function. $$f(x)=x^2$$ So, I want to use this function as the base function.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 100)
f = x**2

plt.plot(x, f)
plt.show()
```

### Step2: Make it integrable

As a distribution function, it must be a finite function and integrable. Let's take it up side down by multiplying minus.$$f(x)=-x^2$$

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 100)
f = -x**2

plt.plot(x, f)
plt.show()
```

### Step3: Make a bell curve

The function is still a bit strange since it slopes down suddenly. To make it smooth, I want to use natural base function ($y=e^n$).

Then, our function is going to be $$f(x)=e^{-x^2}$$

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 100)
f = np.exp(-x**2)

plt.plot(x, f)
plt.show()
```

### Step4: Probability distribution

Our function area must be 1 to be a distribution function since the sum of every probability should be 100%. 

Gaussian integral defines the following.

$$
\int_{-\infty}^{\infty} e^{-a(x+b)^2} dx = \sqrt{\frac{\pi}{a}}
$$

Let $y$ be $x = y/\sqrt{2}$ then our function is

$$
f(y) = e^{-\frac{1}{2}y^2}
$$

The integral of the function should be

$$
\int_{-\infty}^{\infty} e^{-\frac{1}{2}y^2} dy = \sqrt{2\pi}
$$

deviding both side by $\sqrt{2\pi}$

$$
\int_{-\infty}^{\infty} \frac{1}{\sqrt{2\pi}}e^{-\frac{1}{2}y^2} dy = 1
$$

It's same with the standard normal distribution!

$$
\int_{-\infty}^{\infty} \frac{1}{\sqrt{2\pi}}\exp(-\frac{1}{2}y^2) dy = 1
$$


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 100)
f = (1/np.sqrt(2*np.pi))*np.exp(-x**2/2)

plt.plot(x, f)
plt.show()
```
